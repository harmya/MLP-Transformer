```python
import json
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
embed_path = "embeddings.json"
with open(embed_path, "r") as f:
    embeddings = json.load(f)
```

```python
sentence = "laughter joy sad fear"
```

```python
sentence_embedding = [np.array(embeddings[word]) for word in sentence.split() if word in embeddings]
sentence_embedding = np.array(sentence_embedding)
```

```python
Q = sentence_embedding
K = sentence_embedding.T
```

```python
qkt = np.dot(Q, K)
qkt = qkt - np.diag(np.diag(qkt))
```

```python
plt.imshow(qkt, cmap="binary", interpolation="nearest")
plt.colorbar()
plt.xticks(range(len(sentence.split())), sentence.split())
plt.yticks(range(len(sentence.split())), sentence.split())
plt.title("QK^T")
plt.xlabel("Q")
plt.ylabel("Kt")
plt.show()
```
